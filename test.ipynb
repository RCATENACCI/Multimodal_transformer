```python
from pydub import AudioSegment

def keep_first_10_seconds(input_path, output_path):

    audio = AudioSegment.from_file(input_path)

    first_10_seconds = audio[:10000]  # 10000 ms

    first_10_seconds.export(output_path, format="mp3")

audio_path = "data/$NVDA NVIDIA Q1 2026 Earnings Conference Call [PzrX70OYhDo].webm"

keep_first_10_seconds(
    r"C:\Users\RaphaelCatenacci\vscodeprojects\Multimodal_transformer\data\$NVDA NVIDIA Q1 2026 Earnings Conference Call [PzrX70OYhDo].webm",
    "data/nvidia_earnings_short.mp3"
)
```
